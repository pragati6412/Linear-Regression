```python
import pandas as pd
import numpy as np
import matplotlib.axes as ax
import matplotlib.pyplot as plt
df=pd.read_csv('data_for_lr.csv')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
train_input=np.array(df.x[0:500]).reshape(500,1)
train_output=np.array(df.y[0:500]).reshape(500,1)
valid_input=np.array(df.x[500:700]).reshape(199,1)
valid_output=np.array(df.y[500:700]).reshape(199,1)
```

```python
lr_model=LinearRegression()
lr_model.fit(train_input,train_output)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

```python
test_prediction=lr_model.predict(valid_input)
```

```python
plt.plot(valid_input,test_prediction,"+",color="black")
plt.plot(valid_input,valid_output,"*",color="red")
plt.title("Performance Testing")
plt.xlabel("Input")
plt.ylabel("Output/Prediction")
```

```python
from sklearn.metrics import mean_squared_error
cost=mean_squared_error(valid_output,test_prediction)
```
